```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read Dataset
data = pd.read_csv('kc_house_data - copy.csv')
data.head()
```

```python
data.info()
```

```python
# dropped float type features
data = data.drop(['date', 'bathrooms', 'floors', 'lat', 'long'], axis =1)
data.head()
```

```python
# Extract Features and Target
X = data.drop(['price'], axis=1)
Y = data['price']
```

# Scaled using mimmax scaler

```python
# Using minmax scaler to scale 'X' features
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler() 
scaled_data = scaler.fit_transform(X)
scaled_data                                 
```

```python
# Convert the scaled features to a DataFrame
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data
```

# Scaled using minmax formula

```python
# Calculate the minimum and maximum 
X_min = X.min()
X_max = X.max()

# Apply Min-Max Scaling
X_scaled = (X - X_min) / (X_max - X_min)
```
